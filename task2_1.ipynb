```python
import pandas as pd
import numpy as np

path = "/content/drive/MyDrive/itmo_ml/24_25.csv"
table = pd.read_csv(path, header=None)   # без заголовка
print(table.shape)                       # (60, 10)
display(table.head())
```

```python
from sklearn.decomposition import PCA

pca = PCA()                  # берём все 10 компонент
X_pca = pca.fit_transform(table.values)
```

```python
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_explained = np.cumsum(explained_variance_ratio)

print(explained_variance_ratio)
print(cumulative_explained)
```

```python
pc1_obj1 = X_pca[0, 0]
pc2_obj1 = X_pca[0, 1]

pc1_obj1_round = round(pc1_obj1, 3)
pc2_obj1_round = round(pc2_obj1, 3)
print(pc1_obj1)
print(pc2_obj1)
print(pc1_obj1_round)
print(pc2_obj1_round)
```

```python
first_two_var = explained_variance_ratio[0] + explained_variance_ratio[1]
first_two_var_round = round(first_two_var, 3)
print(first_two_var_round)
```

```python
min_components_085 = int(np.argmax(cumulative_explained > 0.85) + 1)
print(min_components_085)
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.scatter(X_pca[:, 0], X_pca[:, 1])
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("Objects in the space of first two principal components")
plt.grid(True)
plt.show()
```
